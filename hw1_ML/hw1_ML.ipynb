```python
# simulate 10000 distances between 2 points of N=2 dimension space.
n=0
dist_list_2dim = list()
while(n<10000):
    x=np.random.rand(1,2)
    y=np.random.rand(1,2)
    dist=np.linalg.norm(x-y)  
    dist_list_2dim.append(dist)
    n = n+1

mean_2dim = np.mean(dist_list_2dim)
var_2dim = np.var(dist_list_2dim)
skew_2dim = stats.skew(dist_list_2dim)
kurtosis_2dim = stats.kurtosis(dist_list_2dim)

print("mean, variance, skewness, kurtosis of N=2: ")
print(mean_2dim)
print(var_2dim)
print(skew_2dim)
print(kurtosis_2dim)
```

question2: linear regression for all stocks:

```python
%pylab inline
import os.path
import pandas as pd
from pymlfin.util.util import get_eps_data

import statsmodels.formula.api as sm
import statsmodels.api as sm_api
import sklearn
import sklearn.model_selection
import sklearn.linear_model
import matplotlib.dates

print(os.getcwd())

all_stock_list = ['HON','JPM','XOM','HAL','HOV','INTC','IBM','L','MDT','MSFT']
for stock in all_stock_list:
    df = get_eps_data(stock)
    df.reset_index(level=0, inplace=True) # remove GVKEY
    df.loc[:,('EPS_t1', 'EPS_t4', 'INV_t1', 'AR_t1', 'SA_t1', 'GM_t1', 'CAPX_t1', 'ETR_t1', 'EPS')].head()

    # M2.1 by Lorek and Willinger (1996)
    # E(EPS[t]) = alpha + beta_1 *EPS[t-1] + beta_2*EPS[t-4] +
    # beta_3 * INV[t-1] + beta_4 * AR[t-1] + beta_5 * CAPX[t-1] + beta_6 *GM[t-1] +
    # beta_7 * SA[t-1] + beta_8 * ETR[t-1] + beta9 * LF[t-1] + epsilon_t

    eps_df = pd.DataFrame(df['EPS'].values, columns=['EPS'], index=df.index)
    predictors_df = df.loc[:, ('EPS_t1', 'EPS_t4', 'INV_t1', 'AR_t1', 'CAPX_t1', 'GM_t1', 'SA_t1', 'ETR_t1')]

    n_splits = 2
    tscv = sklearn.model_selection.TimeSeriesSplit(n_splits=n_splits)
    for train_index, test_index in tscv.split(predictors_df):
        predict_train, predict_test = predictors_df.loc[predictors_df.index[train_index]], \
                                      predictors_df.loc[predictors_df.index[test_index]]
        eps_train, eps_test = eps_df.loc[eps_df.index[train_index]], \
                              eps_df.loc[eps_df.index[test_index]]

        lm_lorek = sklearn.linear_model.LinearRegression().fit(predict_train, eps_train)  # train data to fit model
        # each stock would yield two regression results of different splitting sample size.
        print("-------------------------Resulting for company: %s" % stock +"--------------------------")
        print("M2.1 Lorek and Willinger (1996) via scikit-learn")
        print("training data set has %d observations from %s to %s" % (len(predict_train.index), 
                                                                       predict_train.index[0].strftime('%Y-%m-%d'),
                                                                       predict_train.index[-1].strftime('%Y-%m-%d')))
        print("model coefficients: %s" % lm_lorek.coef_)
        print("model intercept: %.4f" % lm_lorek.intercept_[0])

        rmse = np.sqrt(sklearn.metrics.mean_squared_error(eps_train, lm_lorek.predict(predict_train)))
        print("scikit-learn in-sample RMSE = %.4f" % rmse)

        eps_forecast = lm_lorek.predict(predict_test)
        rmse = np.sqrt(sklearn.metrics.mean_squared_error(eps_test, eps_forecast))
        print("scikit-learn out-of-sample RMSE = %.4f" % rmse)
        eps_forecast = pd.DataFrame(eps_forecast, index=eps_test.index, columns=('EPS Forecast',))
        comb = pd.concat([eps_forecast, eps_test], axis=1, join='inner')
        comb.rename(columns={0: 'EPS Forecast'}, inplace=True)
        comb.describe()
        ax = comb.plot(title='Out-of-sample M2.1 Lorek and Willinger (1996) via sklearn.linear_model.LinearRegression', 
                       figsize=(16, 8))
```

Question3: L2 regularization: use Ridge regression. 

```python
%pylab inline
import os.path
import pandas as pd
from pymlfin.util.util import get_eps_data

print(os.getcwd())
df = get_eps_data('MSFT')
df.reset_index(level=0, inplace=True) # remove GVKEY
df.loc[:,('EPS_t1', 'EPS_t4', 'INV_t1', 'AR_t1', 'SA_t1', 'GM_t1', 'CAPX_t1', 'ETR_t1', 'EPS')].head()

# M2.1 by Lorek and Willinger (1996)
# E(EPS[t]) = alpha + beta_1 *EPS[t-1] + beta_2*EPS[t-4] +
# beta_3 * INV[t-1] + beta_4 * AR[t-1] + beta_5 * CAPX[t-1] + beta_6 *GM[t-1] +
# beta_7 * SA[t-1] + beta_8 * ETR[t-1] + beta9 * LF[t-1] + epsilon_t

eps_df = pd.DataFrame(df['EPS'].values, columns=['EPS'], index=df.index)
predictors_df = df.loc[:, ('EPS_t1', 'EPS_t4', 'INV_t1', 'AR_t1', 'CAPX_t1', 'GM_t1', 'SA_t1', 'ETR_t1')]

import statsmodels.formula.api as sm
import statsmodels.api as sm_api
import sklearn
import sklearn.model_selection
import sklearn.linear_model
import matplotlib.pyplot as plt

n_splits = 2
tscv = sklearn.model_selection.TimeSeriesSplit(n_splits=n_splits)

alpha_list = [0.01, 0.1, 0.2, 0.5, 0.8, 1.0, 1.2, 1.5, 2.0]
rmse2_list = []  
rmse_finalResult_list = []

for a in alpha_list:
    
    for train_index, test_index in tscv.split(predictors_df):
        predict_train, predict_test = predictors_df.loc[predictors_df.index[train_index]], \
                                      predictors_df.loc[predictors_df.index[test_index]]
        eps_train, eps_test = eps_df.loc[eps_df.index[train_index]], \
                              eps_df.loc[eps_df.index[test_index]]

        lm_lorek = sklearn.linear_model.Ridge(alpha=a).fit(predict_train, eps_train)  # train data to fit model
        #print("M2.1 Lorek and Willinger (1996) via scikit-learn")
        #print("training data set has %d observations from %s to %s" % (len(predict_train.index), predict_train.index[0].strftime('%Y-%m-%d'),predict_train.index[-1].strftime('%Y-%m-%d')))
        #print("model coefficients: %s" % lm_lorek.coef_)
        #print("model intercept: %.4f" % lm_lorek.intercept_[0])

        #rmse1 = np.sqrt(sklearn.metrics.mean_squared_error(eps_train, lm_lorek.predict(predict_train)))
        #print("scikit-learn in-sample RMSE = %.4f" % rmse1)

        eps_forecast = lm_lorek.predict(predict_test)
        rmse2 = np.sqrt(sklearn.metrics.mean_squared_error(eps_test, eps_forecast))
        #print("scikit-learn out-of-sample RMSE = %.4f" % rmse2)
        rmse2_list.append(rmse2)
    
print("this is lambda value list: ")
print(alpha_list)
# print(rmse2_list)
# print("-------------")

for i in range(len(rmse2_list)):
    if(i%2 == 0):
        avg_rmse = (rmse2_list[i] + rmse2_list[i+1])/2  # avg. of rmse for each alpha
        rmse_finalResult_list.append(avg_rmse)  # only when i%2==0 do we append

print("corresponding rmse list: ")
print(rmse_finalResult_list)


#eps_forecast = pd.DataFrame(eps_forecast, index=eps_test.index, columns=('EPS Forecast',))
#comb = pd.concat([eps_forecast, eps_test], axis=1, join='inner')
#comb.rename(columns={0: 'EPS Forecast'}, inplace=True)
#comb.describe()
#ax = comb.plot(title='Out-of-sample M2.1 Lorek and Willinger (1996) via sklearn.linear_model.RidgeRegression', figsize=(16, 8))

plt.plot(alpha_list, rmse_finalResult_list)
plt.suptitle('L2 regularization: RMSE-lambda curve')
plt.show()
```

```python
%pylab inline
import os.path
import pandas as pd
from pymlfin.util.util import get_eps_data

print(os.getcwd())
df = get_eps_data('MSFT')
df.reset_index(level=0, inplace=True) # remove GVKEY
df.loc[:,('EPS_t1', 'EPS_t4', 'INV_t1', 'AR_t1', 'SA_t1', 'GM_t1', 'CAPX_t1', 'ETR_t1', 'EPS')].head()

# M2.1 by Lorek and Willinger (1996)
# E(EPS[t]) = alpha + beta_1 *EPS[t-1] + beta_2*EPS[t-4] +
# beta_3 * INV[t-1] + beta_4 * AR[t-1] + beta_5 * CAPX[t-1] + beta_6 *GM[t-1] +
# beta_7 * SA[t-1] + beta_8 * ETR[t-1] + beta9 * LF[t-1] + epsilon_t

eps_df = pd.DataFrame(df['EPS'].values, columns=['EPS'], index=df.index)
predictors_df = df.loc[:, ('EPS_t1', 'EPS_t4', 'INV_t1', 'AR_t1', 'CAPX_t1', 'GM_t1', 'SA_t1', 'ETR_t1')]

import statsmodels.formula.api as sm
import statsmodels.api as sm_api
import sklearn
import sklearn.model_selection
import sklearn.linear_model
import matplotlib.pyplot as plt

n_splits = 2
tscv = sklearn.model_selection.TimeSeriesSplit(n_splits=n_splits)

alpha_list = [0.0001, 0.0002, 0.0005, 0.0010, 0.0015, 0.0020, 0.0025, 0.0030, 0.0035, 0.0040, 0.0045, 0.005, 0.0055]
rmse2_list = []  
rmse_finalResult_list = []

for a in alpha_list:
    
    for train_index, test_index in tscv.split(predictors_df):
        predict_train, predict_test = predictors_df.loc[predictors_df.index[train_index]], \
                                      predictors_df.loc[predictors_df.index[test_index]]
        eps_train, eps_test = eps_df.loc[eps_df.index[train_index]], \
                              eps_df.loc[eps_df.index[test_index]]

        lm_lorek = sklearn.linear_model.Lasso(alpha=a).fit(predict_train, eps_train)  # train data to fit model
        
        eps_forecast = lm_lorek.predict(predict_test)
        rmse2 = np.sqrt(sklearn.metrics.mean_squared_error(eps_test, eps_forecast))
        #print("scikit-learn out-of-sample RMSE = %.4f" % rmse2)
        rmse2_list.append(rmse2)
    
print("this is lambda value list: ")
print(alpha_list)

for i in range(len(rmse2_list)):
    if(i%2 == 0):
        avg_rmse = (rmse2_list[i] + rmse2_list[i+1])/2  
        rmse_finalResult_list.append(avg_rmse) 

print("corresponding rmse list: ")
print(rmse_finalResult_list)

plt.plot(alpha_list, rmse_finalResult_list)
plt.suptitle('L1 regularization: RMSE-lambda curve')
plt.show()
```
